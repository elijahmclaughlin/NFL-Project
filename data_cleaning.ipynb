```python
import pandas as pd
```

```python
offense = pd.read_csv("offense.csv")
```

```python
offense = offense[["Tm", "Yds", "Ply", "Y/P", "TO", "FL", "1stD", "Pen", "Yds.3", "1stPy", "Sc%", "TO%", "Year"]]
```

```python
offense.rename(columns={'Yds': 'Total Yards'}, inplace=True)
```

```python
offense.rename(columns={'Yds.3': 'Penalty Yards'}, inplace=True)
```

```python
offense.rename(columns={'EXP': 'Total EXP'}, inplace=True)
```

```python
team = pd.read_csv("team.csv")
```

```python
del team["Unnamed: 0"]
```

```python
team["Tm"] = team["Tm"].str.replace("*", "", regex=False)
```

```python
team["Tm"] = team["Tm"].str.replace("+", "", regex=False)
```

```python
combined_team = team.merge(offense, how="outer", on=["Tm", "Year"])
```

```python
passing = pd.read_csv("passing.csv")
```

```python
passing = passing[["Tm", "Cmp", "Att", "Cmp%", "Yds", "TD", "Int", "Y/A", "Y/G", "Sk", "EXP", "Year"]]
```

```python
passing.rename(columns={'Yds': 'Passing Yards'}, inplace=True)
```

```python
passing.rename(columns={'Y/A': 'Passing Y/A'}, inplace=True)
```

```python
passing.rename(columns={'Y/G': 'Passing Y/G'}, inplace=True)
```

```python
passing.rename(columns={'EXP': 'Passing EXP'}, inplace=True)
```

```python
passing.rename(columns={'Att': 'Passing Att'}, inplace=True)
```

```python
passing.rename(columns={'TD': 'Passing TD'}, inplace=True)
```

```python
rushing = pd.read_csv("rushing.csv")
```

```python
del rushing["Unnamed: 0"]
```

```python
del rushing["Rk"]
```

```python
del rushing["G"]
```

```python
del rushing["Lng"]
```

```python
rushing.rename(columns={'Yds': 'Rushing Yards'}, inplace=True)
```

```python
rushing.rename(columns={'Att': 'Rushing Att'}, inplace=True)
```

```python
rushing.rename(columns={'Y/A': 'Rushing Y/A'}, inplace=True)
```

```python
rushing.rename(columns={'Y/G': 'Rushing Y/G'}, inplace=True)
```

```python
rushing.rename(columns={'EXP': 'Rushing EXP'}, inplace=True)
```

```python
rushing.rename(columns={'TD': 'Rushing TD'}, inplace=True)
```

```python
combined_pr = passing.merge(rushing, how="outer", on=["Tm", "Year"])
```

```python
stats = combined_team.merge(combined_pr, how="outer", on=["Tm", "Year"])
```

```python
stats['Tm'] = stats['Tm'].replace(['Washington Redskins', 'Washington Football Team'], 'Washington Commanders')
```

```python
stats['Tm'] = stats['Tm'].replace(['St. Louis Rams'], 'Los Angeles Rams')
```

```python
stats['Tm'] = stats['Tm'].replace(['San Diego Chargers'], 'Los Angeles Chargers')
```

```python
stats['Tm'] = stats['Tm'].replace(['Oakland Raiders'], 'Las Vegas Raiders')
```

```python
stats.to_csv("team_stats.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
highest_scoring = stats.sort_values("PF", ascending=False).head(10)
```

```python
ax = highest_scoring.plot.bar(x='Tm', y='PF', legend=False)

for i, row in highest_scoring.iterrows():
    team = row['Tm']
    year = stats.loc[(stats['Tm'] == team), 'Year'].values[0]
    ax.text(i, row['PF'], year, ha='center')

ax.set_xlabel('Team')
ax.set_ylabel('Points For')
ax.set_title('Top 10 Highest Scoring Offenses')

plt.show()
```

```python
worst_def = stats.sort_values("PA", ascending=False).head(10)
```

```python
ax1 = worst_def.plot.bar(x='Tm', y='PA', legend=False)

for i, row in worst_def.iterrows():
    team = row['Tm']
    year = row['Year']
    ax.text(i, row['PA'], year, ha='center')

ax1.set_xlabel('Team')
ax1.set_ylabel('Points Against')
ax1.set_title('Top 10 Worst Defenses')

plt.show()
```
